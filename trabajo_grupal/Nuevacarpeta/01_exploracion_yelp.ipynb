## Conteo de reviews por aÃ±o usando chunks

Como el archivo de reviews es muy grande, lo leemos por partes. 
Primero contamos cuÃ¡ntas reviews hay por aÃ±o para elegir un filtro temporal razonable.


```python
from collections import Counter

conteo_por_anio = Counter()
chunksize = 100_000

for numero_chunk, chunk in enumerate(pd.read_json(review_path, lines=True, chunksize=chunksize), start=1):
    anios = chunk["date"].dt.year
    conteo_por_anio.update(anios.value_counts().to_dict())
    
    if numero_chunk % 10 == 0:
        print(f"Chunks procesados: {numero_chunk} | filas aproximadas: {numero_chunk * chunksize:,}")

df_conteo_anios = (
    pd.DataFrame.from_dict(conteo_por_anio, orient="index", columns=["cantidad_reviews"])
    .sort_index()
    .rename_axis("anio")
    .reset_index()
)

df_conteo_anios
```

## Filtrado de reviews en 2021

Elegimos este aÃ±o porque es el Ãºltimo aÃ±o completo y cumple con 618 189 filas (>500)

```python
reviews_2021 = []
chunksize = 100_000

for numero_chunk, chunk in enumerate(pd.read_json(review_path, lines=True, chunksize=chunksize), start=1):
    chunk_2021 = chunk[(chunk["date"] >= "2021-01-01") & (chunk["date"] < "2022-01-01")]
    reviews_2021.append(chunk_2021)
    
    if numero_chunk % 10 == 0:
        print(f"Chunks procesados: {numero_chunk}")

df_reviews_2021 = pd.concat(reviews_2021, ignore_index=True)

df_reviews_2021.shape
```
